```python
import pandas as pd

daily_activity = pd.read_csv("dailyActivity_merged.csv")
sleep_day = pd.read_csv("sleepDay_merged.csv")
Intensities = pd.read_csv("hourlyIntensities_merged.csv")
daily_calories = pd.read_csv("dailyCalories_merged.csv")
weight_info = pd.read_csv("weightLogInfo_merged.csv")
```

```python
print(daily_activity.head())
print(daily_calories.head())
```

```python
# Split the 'SleepDay' column into 'Date' and 'Time' columns
sleep_day['Date'], sleep_day['Time'] = zip(*sleep_day['SleepDay'].apply(lambda x: x.split(' ', 1)))

# Split the 'Date' column into 'date' and 'time' columns
weight_info['date'], weight_info['time'] = zip(*weight_info['Date'].apply(lambda x: x.split(' ', 1)))

# Split the 'ActivityHour' column into 'Date' and 'Time' columns
Intensities['Date'], Intensities['Time'] = zip(*Intensities['ActivityHour'].apply(lambda x: x.split(' ', 1)))

# Print the first few rows of the dataframes
print(sleep_day.head())
print(weight_info.head())
print(Intensities.head())
```

```python


def clean_names(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    return df

# Call the clean_names function on your dataframes
daily_activity = clean_names(daily_activity)
daily_calories = clean_names(daily_calories)
Intensities = clean_names(Intensities)
sleep_day = clean_names(sleep_day)
weight_info = clean_names(weight_info)
```

```python
daily_activity_cols = daily_activity.columns
daily_calories_cols = daily_calories.columns
Intensities_cols = Intensities.columns
sleep_day_cols = sleep_day.columns
weight_info_cols = weight_info.columns

daily_activity_cols = list(daily_activity_cols)
daily_calories_cols = list(daily_calories_cols)
Intensities_cols = list(Intensities_cols)
sleep_day_cols = list(sleep_day_cols)
weight_info_cols = list(weight_info_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
plt.scatter(daily_activity['totalsteps'], daily_activity['calories'])

# Add a trend line
sns.regplot(x=daily_activity['totalsteps'], y=daily_activity['calories'], scatter=False, truncate=True)

# Set labels for the axes and the title
plt.xlabel('Total Steps')
plt.ylabel('Calories')
plt.title('Total Steps vs Calories')

# Show the plot
plt.show()
```
